```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

Name = []
Prices = []

for i in range(1,15):

    url = "https://www.amazon.in/s?k=mobiles+under+10000&crid=2MWQ9EMZYOR4P&qid=1718430757&sprefix=mobile+%2Cstripbooks%2C368&ref=sr_pg_" + str(i)

    header = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36" ,'referer':'https://www.google.com/'}

    r = requests.get(url,headers = header)
    print(r)
    soup = BeautifulSoup(r.text, "lxml")
# print(soup)
    # np = soup.find("a",class_ = "s-pagination-item").get("href") # This line seems unrelated to the issue and can be removed
    # print("https://www.amazon.in"+ np)

    box = soup.find("div",class_ = "s-main-slot")

    names = box.find_all("h2", class_ = "a-size-mini")
    prices = box.find_all("span", class_ = "a-price-whole")

    # Ensure that names and prices are extracted in pairs
    for name, price in zip(names, prices):
        Name.append(name.text.strip()) # Strip any extra spaces
        Prices.append(price.text.strip())

print(Name)
print(Prices)

df = pd.DataFrame({"Product Name": Name, "Price List": Prices})
print(df)
```

```python
df.to_csv("Amazon_Mobile_Prices.csv")
```
